```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Creating a Class of various Activation Functions**

```python
class ActivationFunction:
    def __init__(self, function_name):
        self.function_name = function_name
    
    def apply_activation(self,x):
        if self.function_name == "relu":
            return np.maximum(0,x)
        elif self.function_name == "linear":
            return x
        elif self.function_name == "sigmoid":
            return 1/(1 + np.exp(-x))
        elif self.function_name == "tanh":
            return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)) #can be done using np.tanh(x) as well
        elif self.function_name == "softmax":
            return np.exp(x) / np.sum(np.exp(x),axis=0)
        else:
            return None
```

**1 input, 1 Layer Neural Network [1,1,1]**

```python
x = [-2,3]              #input features
w = [0.5,0.8]           #weights
b = 1                   #bias
y = np.dot(x,w) + b     #output
```

```python
############################################
# Creating Objects for each Activation Function
############################################
activation_relu = ActivationFunction("relu")
activation_linear = ActivationFunction("linear")
activation_sigmoid = ActivationFunction("sigmoid")
activation_tanh = ActivationFunction("tanh")
activation_softmax = ActivationFunction("softmax")

############################################
# Applying Activation Functions
############################################
z_relu = activation_relu.apply_activation(y)
z_linear = activation_linear.apply_activation(y)
z_sigmoid = activation_sigmoid.apply_activation(y)
z_tanh = activation_tanh.apply_activation(y)
z_softmax = activation_softmax.apply_activation(y)

print(f"Output through ReLu : {z_relu}")
print(f"Output through Linear : {z_linear}")
print(f"Output through Sigmoid : {z_sigmoid}")
print(f"Output through Tanh : {z_tanh}")
print(f"Output through Softmax : {z_softmax}")
```

**1 input, 2 Layer Neural Network [1,2,1]**

```python
x = [-2,3]
w1 = [0.5,0.8]
w2 = [0.2,0.3]
b = 1
y1 = np.dot(x,w1) + np.dot(x,w2) + b

############################################
# Applying Activation Functions
############################################
z_relu1 = activation_relu.apply_activation(y1)
z_linear1 = activation_linear.apply_activation(y1)
z_sigmoid1 = activation_sigmoid.apply_activation(y1)
z_tanh1 = activation_tanh.apply_activation(y1)
z_softmax1 = activation_softmax.apply_activation(y1)

print(f"Output through ReLu : {z_relu1}")
print(f"Output through Linear : {z_linear1}")
print(f"Output through Sigmoid : {z_sigmoid1}")
print(f"Output through Tanh : {z_tanh1}")
print(f"Output through Softmax : {z_softmax1}")
```

**2 input, 2 layer Neural Network [2,2,1]**

```python
x1 = [-2,3]
x2 = [1,4]
w1 = [0.5,0.8]
w2 = [0.2,0.3]
b = 1
y2 = np.dot(x1,w1) + np.dot(x2,w2) + np.dot(x1,w2) + np.dot(x2,w1) + b

############################################
# Applying Activation Functions
############################################
z_relu2 = activation_relu.apply_activation(y2)
z_linear2 = activation_linear.apply_activation(y2)
z_sigmoid2 = activation_sigmoid.apply_activation(y2)
z_tanh2 = activation_tanh.apply_activation(y2)
z_softmax2 = activation_softmax.apply_activation(y2)

print(f"Output through ReLu : {z_relu2}")
print(f"Output through Linear : {z_linear2}")
print(f"Output through Sigmoid : {z_sigmoid2}")
print(f"Output through Tanh : {z_tanh2}")
print(f"Output through Softmax : {z_softmax2}")
```

**Plotting the Activation Functions**

```python
def apply_plot(x,activation_function):
    y = activation_function.apply_activation(x)
    plt.figure(figsize=(5,5))
    plt.plot(x,y)
    plt.title(f"{activation_function.function_name} function")
    plt.xlabel("x")
    plt.ylabel(f"{activation_function.function_name}(x)")
    plt.grid()
    plt.show()

x = np.linspace(-10,10,100)

apply_plot(x,activation_relu)
apply_plot(x,activation_linear)
apply_plot(x,activation_sigmoid)
apply_plot(x,activation_tanh)
apply_plot(x,activation_softmax)
```
